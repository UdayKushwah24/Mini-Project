```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
import pickle
from sklearn.metrics import r2_score

# Load dataset
data = pd.read_csv("House_Price_Prediction_Dataset.csv")

# Encode categorical columns
label_cols = ['Location', 'Condition', 'Garage']
encoder = LabelEncoder()
for col in label_cols:
    data[col] = encoder.fit_transform(data[col])

# Features and Target
X = data.drop(['Id', 'Price'], axis=1)
y = data['Price']

# Train model on full data  (No Train-Test Split since dataset is small)
model = RandomForestRegressor(n_estimators=500, random_state=42)
model.fit(X, y)

# Model Accuracy (Train accuracy check)
train_score = model.score(X, y)
print("Training Accuracy (R²):", train_score)

# Save model
pickle.dump(model, open("model.pkl", "wb"))
pickle.dump(encoder, open("encoder.pkl", "wb"))
print("✅ Model Trained & Saved Successfully!")
```
